# One-Hot Encoding

Una vez eliminados los valores NaN pasamos a codificar valores con one-hot para pasar algunas variables no numericas a numericas para poderlas utilizar. En este caso serán club, preferred foot y position. De esta forma, diferenciaremos cada una de las opciones definidas en estas variables en variables independientes y las que contuviera el jugador tendrán valor 1 y las que no contuviera el jugador tendrán valor 0. También lo podriamos utilizar para otras variables como la nacionalidad, pero desde mi punto de vista no tiene relación con el valor de un jugador, ya que ser de X país no aumenta tu valor, en principio solo tu habilidad.

```python
clb = df.pop("Club")

df = pd.concat([df.reset_index(drop=True), pd.get_dummies(clb, prefix='clb').reset_index(drop=True)], axis=1, sort=False)

pfoot = df.pop("Preferred Foot")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pfoot, prefix='pfoot').reset_index(drop=True)], axis=1, sort=False)

pos = df.pop("Position")
df = pd.concat([df.reset_index(drop=True), pd.get_dummies(pos, prefix='pos').reset_index(drop=True)], axis=1, sort=False)
df.loc [:2, ["Name", "clb_FC Barcelona", "clb_Juventus", "clb_Paris Saint-Germain","pfoot_Right","pfoot_Left", "pos_RF","pos_ST","pos_LF"]]
```

# Conversión de Valores

Ahora que hemos terminado con el onehot pasaremos a convertir valores en floats como es el caso de valor (variable a predecir), el salario o la clausula de rescisión.

Para ello, utilizaremos la clase value_to_float que se nos dio en la introducción de la práctica. De este modo quitaremos el simbolo € y haremos la conversion de K(mil) y M(millones) en las variables que hemos mencionado.

```python
def value_to_float(x):
    """
    From K and M to float.
    
    """
    x = x.replace('€', '')
    ret_val = 0.0
    
    if type(x) == float or type(x) == int:
        ret_val = x
    if 'K' in x:
        if len(x) > 1:
            ret_val = float(x.replace('K', ''))
        ret_val = ret_val *1000
    if 'M' in x:
        if len(x) > 1:
            ret_val = float(x.replace('M', ''))
        ret_val = ret_val * 1000000.0
    return ret_val
```

Ahora aplicaremos esta funcion a las 3 variables antes mencionadas.

```python
df["Value"] = df["Value"].apply(value_to_float)
df["Release Clause"] = df["Release Clause"].apply(value_to_float)
df["Wage"] = df["Wage"].apply(value_to_float)
df
```

Ahora pasaremos al caso de peso y la altura, por un lado el peso podemos unicamente pasarlo a un valor float sin necesidad de convertir las lbs en kg

Pero en el caso de la altura no es realista ya que pasarlo estrictamente a un falor float es erroneo, por lo que hay que convertir los pies y pulgadas en cm.
Primero, veamos la conversión del peso:

```python
df["Weight"] = df["Weight"].str.replace('lbs','.0')
df["Weight"] = pd.to_numeric(df["Weight"], downcast="float")
```

Y para el caso de la altura hay que pasarlo del sistema imperial al metrico. En este caso 1 pie = 30,48cm y 1 pulgada = 2,54cm.
Por lo tanto realizaremos un clase que sirva de conversor. Utilizaremos split con el simbolo ' para separar la altura imperial en 2 (pies y pulgadas) y luego realizaremos la conversion con los valores que hemos indicado antes

```python
def height_in_metric(ht):
    altura = ht.split("'")
    pies = float(altura[0])
    pulgadas = float(altura[1])
    return int (30.48 * pies + 2.54 * pulgadas)
```

Ahora utilizaremos esta conversión para convertir todos los valores de la columna Height de imperial al sistema metrico.

```python
df["Height"] = df["Height"].apply(height_in_metric)
df.loc[:10, ["Name","Height", "Weight"]]
```

Los valores economicos como la clausula de rescisión y el salario han supuesto tambien muy imporantes para un mejor rendimiento de las predicciones

```python
val = df.pop("Value")

TrainingData = ["Overall", "Potential", "Wage","International Reputation","Height", "Weight", "Release Clause"]
to_drop = [c for c in df.columns if not c.startswith("clb") and not c.startswith("pos") and not c in TrainingData]

df = df.drop(to_drop, axis=1)

df
```

Ahora que tenemos en nuestro dataframe todos los datos que vamos a utilizar realizamos la separación entre los datos de entrenamiento y los de test.

```python
X_train, X_test, y_train, y_test = train_test_split(df, val, test_size=0.33, random_state=42)
```

```python
len(X_train)
```

Ahora pasamos a la fase de entrenamiento de nuestro modelo lineal

```python
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

Finalmente obtenemos la métrica $R^2$ para la regresión, utilizaremos la implementación de <a href="https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html"> sickit-learn</a>.

```python
preds = reg.predict(X_test)
```

```python
preds[0]
```

```python
y_test[0]
```

```python
r2_score(y_test, preds)
```

Este número como se puede comprobar es muy cercano al valor optimo de la métrica $R^2$ que es el valor 1

Ahora realizaremos tambien un modelo de Stochastic Gradien Descent (SGD) de regresión

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

reg = make_pipeline(StandardScaler(), SGDRegressor(max_iter=1000, tol=1e-5))
reg.fit(X_train, y_train)
```

```python
preds = reg.predict(X_test)
```

```python
preds[0]
```

```python
y_test[0]
```

```python
r2_score(y_test, preds)
```

El dato obtenido también es muy bueno, pero podemos comprobar que el resultado obtenido con la regresión lineal es mejor.
